```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow import keras
from tensorflow.keras import layers


# ========================================================
# 1. Load data
# ========================================================

# 데이터 파일들의 공통 URL 경로 저장
master_url_root = "https://raw.githubusercontent.com/numenta/NAB/master/data/"

# 잡음이 거의 없는 정상 데이터 파일 경로 접미사
df_small_noise_url_suffix = "artificialNoAnomaly/art_daily_small_noise.csv"

# 정상 데이터 전체 URL 생성
df_small_noise_url = master_url_root + df_small_noise_url_suffix

# 정상 데이터를 csv에서 불러오고 timestamp를 인덱스로 설정
df_small_noise = pd.read_csv(
    df_small_noise_url, parse_dates=True, index_col="timestamp"
)

# 이상치가 포함된 데이터 파일 경로 접미사
df_daily_jumpsup_url_suffix = "artificialWithAnomaly/art_daily_jumpsup.csv"

# 이상치 데이터 전체 URL 생성
df_daily_jumpsup_url = master_url_root + df_daily_jumpsup_url_suffix

# 이상치 데이터를 csv에서 불러오기
df_daily_jumpsup = pd.read_csv(
    df_daily_jumpsup_url, parse_dates=True, index_col="timestamp"
)

# 정상 데이터 처음 몇 줄 출력
print(df_small_noise.head())

# 이상치 데이터 처음 몇 줄 출력
print(df_daily_jumpsup.head())

# 정상 데이터 그래프 시각화
fig, ax = plt.subplots()
df_small_noise.plot(legend=False, ax=ax)
plt.show()

# 이상치 데이터 그래프 시각화
fig, ax = plt.subplots()
df_daily_jumpsup.plot(legend=False, ax=ax)
plt.show()


# ========================================================
# 2. Train/Test split + normalization
# ========================================================

# 정상 데이터의 평균값 계산
training_mean = df_small_noise.mean()

# 정상 데이터의 표준편차 계산
training_std = df_small_noise.std()

# 정상 데이터를 평균과 표준편차로 정규화
df_training_value = (df_small_noise - training_mean) / training_std

# 학습 샘플 개수 출력
print("Number of training samples:", len(df_training_value))

# ========================================================
# 3. Sequence generator
# ========================================================

# 시퀀스 길이 정의
TIME_STEPS = 576

# 입력 데이터를 일정 길이 시퀀스로 나누는 함수
def create_sequences(values, time_steps=TIME_STEPS):
    # 결과를 담을 리스트 생성
    output = []

    # 전체 데이터에서 time_steps 만큼씩 슬라이딩하며 시퀀스 생성
    for i in range(len(values) - time_steps + 1):
        output.append(values[i : (i + time_steps)])

    # 리스트를 배열로 변환하여 반환
    return np.stack(output)

# 학습 데이터를 시퀀스로 변환
x_train = create_sequences(df_training_value.values)

# 변환된 학습 데이터의 shape 출력
print("Training input shape: ", x_train.shape)

# ========================================================
# 4. Autoencoder model
# ========================================================

# Conv1D 기반 Autoencoder 모델 구조 정의
model = keras.Sequential(
    [
        # 입력 레이어 정의
        layers.Input(shape=(x_train.shape[1], x_train.shape[2])),

        # 첫번째 인코더 Conv1D 레이어
        layers.Conv1D(
            filters=32,
            kernel_size=7,
            padding="same",
            strides=2,
            activation="relu",
        ),

        # dropout 적용
        layers.Dropout(rate=0.2),

        # 두번째 인코더 Conv1D 레이어
        layers.Conv1D(
            filters=16,
            kernel_size=7,
            padding="same",
            strides=2,
            activation="relu",
        ),

        # 첫번째 디코더 Conv1DTranspose 레이어
        layers.Conv1DTranspose(
            filters=16,
            kernel_size=7,
            padding="same",
            strides=2,
            activation="relu",
        ),

        # dropout 적용
        layers.Dropout(rate=0.2),

        # 두번째 디코더 Conv1DTranspose 레이어
        layers.Conv1DTranspose(
            filters=32,
            kernel_size=7,
            padding="same",
            strides=2,
            activation="relu",
        ),

        # 마지막 디코더 레이어 출력 채널은 하나
        layers.Conv1DTranspose(filters=1, kernel_size=7, padding="same"),
    ]
)

# 옵티마이저와 손실함수 설정 후 모델 컴파일
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss="mse")

# 모델 구조 출력
model.summary()

# ========================================================
# 5. Train model
# ========================================================

# 모델 학습 수행
history = model.fit(
    x_train,
    x_train,
    epochs=50,
    batch_size=128,
    validation_split=0.1,
    callbacks=[
        keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, mode="min")
    ],
)

# 학습 손실과 검증 손실 그래프 출력
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.show()

# ========================================================
# 6. Train Reconstruction Error
# ========================================================

# 학습 데이터 복원 결과 예측
x_train_pred = model.predict(x_train)

# 학습 데이터 MAE 손실 계산
train_mae_loss = np.mean(np.abs(x_train_pred - x_train), axis=1)

# MAE 분포 그래프 출력
plt.hist(train_mae_loss, bins=50)
plt.xlabel("Train MAE loss")
plt.ylabel("No of samples")
plt.show()

# 이상 탐지 기준 threshold를 최대 MAE로 설정
threshold = np.max(train_mae_loss)
print("Reconstruction error threshold: ", threshold)

# ========================================================
# 7. Test anomaly detection
# ========================================================

# 첫번째 학습 시퀀스와 그 복원값을 비교하는 그래프
plt.plot(x_train[0])
plt.plot(x_train_pred[0])
plt.show()

# 테스트 데이터를 학습 데이터의 평균과 표준편차로 정규화
df_test_value = (df_daily_jumpsup - training_mean) / training_std

# 테스트 데이터 그래프 출력
fig, ax = plt.subplots()
df_test_value.plot(legend=False, ax=ax)
plt.show()

# 테스트 데이터를 시퀀스로 변환
x_test = create_sequences(df_test_value.values)

# 테스트 데이터 shape 출력
print("Test input shape: ", x_test.shape)

# 테스트 데이터 복원 예측
x_test_pred = model.predict(x_test)

# 테스트 MAE 계산
test_mae_loss = np.mean(np.abs(x_test_pred - x_test), axis=1)

# 1차원 형태로 변환
test_mae_loss = test_mae_loss.reshape((-1))

# 테스트 MAE 분포 그래프 출력
plt.hist(test_mae_loss, bins=50)
plt.xlabel("test MAE loss")
plt.ylabel("No of samples")
plt.show()

# threshold를 초과하는 이상 샘플 탐지
anomalies = test_mae_loss > threshold

# 이상치 개수 출력
print("Number of anomaly samples: ", np.sum(anomalies))

# 이상치 인덱스 출력
print("Indices of anomaly samples: ", np.where(anomalies))

# 실제 데이터 인덱스로 변환한 이상치 위치 저장
anomalous_data_indices = []
for data_idx in range(TIME_STEPS - 1, len(df_test_value) - TIME_STEPS + 1):
    if np.all(anomalies[data_idx - TIME_STEPS + 1 : data_idx]):
        anomalous_data_indices.append(data_idx)

# 이상치 데이터만 추출
df_subset = df_daily_jumpsup.iloc[anomalous_data_indices]

# 전체 데이터와 이상치 구간을 동일 그래프에 표시
fig, ax = plt.subplots()
df_daily_jumpsup.plot(legend=False, ax=ax)
df_subset.plot(legend=False, ax=ax, color="r")
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# 샘플링 레이어 정의 클래스 생성
class Sampling(layers.Layer):

  # call 함수로 샘플링 연산 수행
  def call(self,inputs):

    # 입력값에서 평균과 로그 분산 분리
    z_mean, z_log_var=inputs

    # 배치 크기 계산
    batch=tf.shape(z_mean)[0]

    # 잠재 차원 크기 계산
    dim=tf.shape(z_mean)[1]

    # 표준 정규분포에서 난수 생성
    epsilon= tf.random.normal(shape=(batch,dim))

    # reparameterization trick 수행 후 결과 반환
    return z_mean+tf.exp(0.5*z_log_var)*epsilon



# 잠재 공간 차원 설정
latent_dim=2

# 입력 이미지 크기 설정
encoder_inputs=keras.Input(shape=(28,28,1))

# 첫번째 컨볼루션 층 적용
x=layers.Conv2D(32,3,activation="relu",strides=2,padding="same")(encoder_inputs)

# 두번째 컨볼루션 층 적용
x=layers.Conv2D(64,3,activation="relu",strides=2,padding="same")(x)

# 특성을 일렬로 펼치기
x=layers.Flatten()(x)

# 완전 연결층 적용
x=layers.Dense(64,activation="relu")(x)

# 잠재 평균 생성
z_mean=layers.Dense(latent_dim,name="z_mean")(x)

# 잠재 로그 분산 생성
z_log_var=layers.Dense(latent_dim,name="z_log_var")(x)

# 샘플링 수행
z=Sampling()([z_mean,z_log_var])

# 인코더 모델 구성
encoder=keras.Model(encoder_inputs,[z_mean,z_log_var,z],name="encoder")

# 인코더 구조 출력
encoder.summary()
# 잠재 벡터 입력 정의
latent_inputs=keras.Input(shape=(latent_dim,))

# 완전 연결층으로 차원 확장
x=layers.Dense(7*7*64,activation="relu")(latent_inputs)

# 디코더 입력 형태로 리쉐이프
x=layers.Reshape((7,7,64))(x)

# 첫번째 전치 컨볼루션 층 적용
x=layers.Conv2DTranspose(64,3,activation="relu",strides=2,padding="same")(x)

# 두번째 전치 컨볼루션 층 적용
x=layers.Conv2DTranspose(32,3,activation="relu",strides=2,padding="same")(x)

# 마지막 출력층 생성
decoder_outputs=layers.Conv2DTranspose(1,3,activation="sigmoid",padding="same")(x)

# 디코더 모델 구성
decoder=keras.Model(latent_inputs,decoder_outputs,name="decoder")

# 디코더 구조 출력
decoder.summary()



# VAE 모델 클래스 정의
class VAE(keras.Model):

  # 생성자에서 encoder와 decoder를 입력받음
  def __init__(self,encoder,decoder, **kwargs):
    super().__init__(**kwargs)

    # 인코더 저장
    self.encoder = encoder

    # 디코더 저장
    self.decoder = decoder

    # 전체 손실 기록용 메트릭 생성
    self.total_loss_tracker = keras.metrics.Mean(name="total_loss")

    # 재구성 손실 기록용 메트릭 생성
    self.reconstruction_loss_tracker = keras.metrics.Mean(
        name="reconstruction_loss"
    )

    # KL 손실 기록용 메트릭 생성
    self.kl_loss_tracker= keras.metrics.Mean(name="kl_loss")

  # 모델이 추적하는 메트릭 리스트 정의
  @property
  def metrics(self):
    return[
        self.total_loss_tracker,
        self.reconstruction_loss_tracker,
        self.kl_loss_tracker,
    ]

  # 한 스텝 학습 루틴 정의
  def train_step(self,data):

    # 연산 기록을 위한 그래디언트 테이프 열기
    with tf.GradientTape() as tape:

      # 인코더를 통해 평균, 로그 분산, 샘플링 결과 계산
      z_mean, z_log_var, z=self.encoder(data)

      # 디코더로 재구성 생성
      reconstruction=self.decoder(z)

      # 재구성 손실 계산
      reconstruction_loss=tf.reduce_mean(
          tf.reduce_sum(
              keras.losses.binary_crossentropy(data,reconstruction),axis=(1,2)
          )
      )

      # KL divergence 손실 계산
      kl_loss = -0.5 * (1+z_log_var - tf.square(z_mean)-tf.exp(z_log_var))

      # KL 손실 합산
      kl_loss=tf.reduce_mean(tf.reduce_sum(kl_loss,axis=1))

      # 전체 손실 계산
      total_loss = reconstruction_loss+kl_loss

    # 그래디언트 계산
    grads=tape.gradient(total_loss,self.trainable_weights)

    # 옵티마이저로 그래디언트 적용
    self.optimizer.apply_gradients(zip(grads,self.trainable_weights))

    # 메트릭 업데이트
    self.total_loss_tracker.update_state(total_loss)
    self.reconstruction_loss_tracker.update_state(reconstruction_loss)
    self.kl_loss_tracker.update_state(kl_loss)

    # 기록된 메트릭 반환
    return{
        "loss":self.total_loss_tracker.result(),
        "reconstruction_loss":self.reconstruction_loss_tracker.result(),
        "kl_loss":self.kl_loss_tracker.result()
    }
# MNIST 데이터셋 로드 준비
mnist=keras.datasets.mnist

# MNIST 학습용과 테스트용 데이터 불러오기
(x_train, _),(x_test,_) =mnist.load_data()

# 학습 데이터와 테스트 데이터를 하나로 합치기
mnist_digits=np.concatenate([x_train,x_test],axis=0)

# 차원을 추가하여 모델 입력 형태로 변환하고 정규화 수행
mnist_digits=np.expand_dims(mnist_digits,-1).astype("float32")/255

# VAE 모델 인스턴스 생성
vae= VAE(encoder,decoder)

# 옵티마이저 설정하여 컴파일 수행
vae.compile(optimizer=keras.optimizers.Adam())

# 전체 데이터셋으로 VAE 학습 수행
vae.fit(mnist_digits,epochs=30,batch_size=128)



# 잠재 공간을 그리기 위한 함수 정의
def plot_latent_space(vae,n=10,figsize=15):

  # 출력 이미지의 기본 숫자 크기 설정
  digit_size=28

  # 잠재 공간 검색 범위 설정
  scale=1.0

  # 전체 이미지를 그릴 큰 배열 생성
  figure=np.zeros((digit_size*n,digit_size*n))

  # x축 잠재 값 범위 설정
  grid_x=np.linspace(-scale,scale,n)

  # y축 잠재 값 범위 설정 거꾸로 배치
  grid_y=np.linspace(-scale,scale,n)[::-1]

  # y축 반복 루프 시작
  for i, yi in enumerate(grid_y):

    # x축 반복 루프 시작
    for j, xi in enumerate(grid_x):

      # 잠재 벡터 하나 생성
      z_sample=np.array([[xi,yi]])

      # 디코더로부터 이미지 복원
      x_decoded=vae.decoder.predict(z_sample)

      # 복원된 이미지를 숫자 형태로 변환
      digit=x_decoded[0].reshape(digit_size,digit_size)

      # 결과를 큰 그림 배열에 배치
      figure[
          i*digit_size : (i+1) * digit_size,
          j*digit_size : (j+1) * digit_size,
      ]=digit

  # 시각화용 그림 생성
  plt.figure(figsize=(figsize,figsize))

  # x축 y축 배치용 시작 위치 계산
  start_range=digit_size//2

  # 끝 위치 계산
  end_range=n*digit_size+start_range

  # 눈금 위치 배열 생성
  pixel_range=np.arange(start_range,end_range,digit_size)

  # 눈금 라벨용 x축 값 반올림
  sample_range_x=np.round(grid_x,1)

  # 눈금 라벨용 y축 값 반올림
  sample_range_y=np.round(grid_y,1)

  # x축 눈금 적용
  plt.xticks(pixel_range,sample_range_x)

  # y축 눈금 적용
  plt.yticks(pixel_range,sample_range_y)

  # x축 레이블
  plt.xlabel("z[0]")

  # y축 레이블
  plt.ylabel("z[1]")

  # 전체 그림 출력
  plt.imshow(figure,cmap="Greys_r")

  # 보이기
  plt.show()

# 잠재 공간 그림 함수 실행
plot_latent_space(vae)



# 잠재 공간에 실제 라벨을 표시하는 함수 정의
def plot_label_clusters(vae,data,labels):

  # 인코더를 통해 모든 데이터의 잠재 평균값 추출
  z_mean, _, _=vae.encoder.predict(data)

  # 그림 크기 지정
  plt.figure(figsize=(12,10))

  # 산점도로 잠재 공간 표시
  plt.scatter(z_mean[:,0],z_mean[:,1],c=labels)

  # 색상 바 추가
  plt.colorbar()

  # x축 레이블
  plt.xlabel("z[0]")

  # y축 레이블
  plt.ylabel("z[1]")

  # x축 범위 설정
  plt.xlim(z_mean[:,0].min() - 1, z_mean[:,0].max() + 1)

  # y축 범위 설정
  plt.ylim(z_mean[:,1].min() - 1, z_mean[:,1].max() + 1)

  # 그림 출력
  plt.show()



# MNIST 데이터셋을 다시 라벨 포함하여 불러오기
(x_train,y_train),_=keras.datasets.mnist.load_data()

# 입력 형태로 변환하고 정규화
x_train=np.expand_dims(x_train,-1).astype("float32")/255

# 잠재 공간 라벨 클러스터 시각화 실행
plot_label_clusters(vae,x_train,y_train)



# 학습용 데이터를 정규화하기 위해 255로 나누기
x_train=x_train/255

# 테스트 데이터를 정규화하기 위해 255로 나누기
x_test=x_test/255

# 학습 데이터를 CNN 입력 형태로 변환하기 위한 리쉐이프
x_train=x_train.reshape((len(x_train),28,28,1))

# 테스트 데이터를 CNN 입력 형태로 변환하기 위한 리쉐이프
x_test=x_test.reshape((len(x_test),28,28,1))



#----------실습 2--------------------------------

# 노이즈 강도 설정
noise_factor=0.5

# 원본 학습 데이터를 보관
original_train=x_train

# 학습용 노이즈 생성
noise_train=np.random.normal(0,1,original_train.shape)

# 원본 학습 데이터에 노이즈 더하기
noisy_train=original_train+noise_factor*noise_train

# 테스트 노이즈 생성
noise_test=np.random.normal(0,1,x_test.shape)

# 원본 테스트 데이터에 노이즈 더하기
noisy_test=x_test+ noise_factor* noise_test

#-------------------------------------------------

#--------------------실습 1---------------------------
#------AE 모델 설계------

# 잠재 벡터 차원 수 설정
encoding_dim=2

# Functional API 방식 인코더 입력 정의
input_en=keras.Input(shape=(28,28,1))

# 입력 이미지를 일렬로 펼치기
input_flat_en=keras.layers.Flatten()(input_en)

# 아래 코드는 컨볼루션 레이어를 적용한 인코더 구조 예시
# x=layers.Conv2D(32,3,activation="relu",strides=2,padding="same")(input_en)

# 아래 코드는 두번째 컨볼루션 레이어 예시
# x=layers.Conv2D(64,3,activation="relu",strides=2,padding="same")(x)

# 아래 코드는 컨볼루션 결과를 플래튼 하는 예시
# input_flat_en=keras.layers.Flatten()(x)

# 잠재 공간 인코딩을 위한 dense 레이어
output_en=keras.layers.Dense(encoding_dim,activation='relu')(input_flat_en)

# 인코더 모델 생성
encoder=keras.Model(inputs=input_en,outputs=output_en)



# 디코더 입력 설정
input_de=keras.Input(shape=(encoding_dim,))

# 잠재벡터를 784 차원으로 복원하는 dense 레이어 적용
hidden=keras.layers.Dense(784,activation='relu')(input_de)

# 출력 이미지를 원래 크기인 28 28 1 로 변환
output_de = keras.layers.Reshape((28,28,1))(hidden)
# 컨볼루션 기반 디코더 구조 예시
# hidden=keras.layers.Dense(7*7*64,activation='relu')(input_de)

#  디코더에서 리쉐이프하는 예시
# x=layers.Reshape((7,7,64))(hidden)

# 전치 컨볼루션을 사용한 업샘플링 예시
# x=layers.Conv2DTranspose(64,3,activation="relu",strides=2,padding="same")(x)

# 두번째 전치 컨볼루션 예시
# x=layers.Conv2DTranspose(32,3,activation="relu",strides=2,padding="same")(x)

# 시그모이드 활성화로 이미지 복원하는 마지막 레이어 예시
# decoder_outputs=layers.Conv2DTranspose(1,3,activation="sigmoid",padding="same")(x)

#  컨볼루션 구조 디코더 전체 모델 예시
# decoder=keras.Model(inputs=input_de,outputs=decoder_outputs)

# 디코더 모델 생성
decoder=keras.Model(inputs=input_de,outputs=output_de)


#Sequential API
# encoder = keras.models.Sequential([
#     keras.layers.Dense(encoding_dim,activation='relu',input_shape=(784,))
# ])
# decoder = keras.models.Sequential([
#     keras.layers.Dense(784,activation='relu',input_shape=(2,))
# ])

#-----------------------------------------------------------



# encoder와 decoder를 연결하여 AE 전체 구성
AE=keras.models.Sequential([encoder,decoder])

# 모델 컴파일 mse 손실 함수와 adam 옵티마이저 사용
AE.compile(optimizer='adam',loss='mse')

# 실습1을 수행할 때 사용하는 학습 코드 예시
# history=AE.fit(x_train,x_train,epochs=50, batch_size=256,shuffle=True)

# 노이즈 제거 autoencoder 학습 수행
history=AE.fit(noisy_train,x_train,epochs=50, batch_size=256,shuffle=True)

# 테스트 이미지 복원 결과 생성
decoded_img= AE.predict(x_test)


#----------------실습1-----------------
# n=4
# plt.figure(figsize=(10,6))
# for i in range(1,n+1):
#   ax=plt.subplot(2,n,i)
#   plt.imshow(x_test[i].reshape(28,28), cmap='gray')
#   ax=plt.subplot(2,n,i+n)
#   plt.imshow(decoded_img[i].reshape(28,28),cmap='gray')
# AE 학습 데이터 재구성 오차 계산 예시
# recon_train = AE.evaluate(x_train, x_train, verbose=0)
# recon_test  = AE.evaluate(x_test, x_test, verbose=0)

# print("Train Reconstruction Loss :", recon_train)
# print("Test  Reconstruction Loss :", recon_test)


# reconstruction error 계산 예시
# train_decoded = AE.predict(x_train)
# test_decoded  = AE.predict(x_test)

# train_error = np.mean((x_train - train_decoded)**2, axis=(1,2,3))
# test_error  = np.mean((x_test  - test_decoded )**2, axis=(1,2,3))

# plt.figure(figsize=(10,5))
# plt.hist(train_error, bins=50, alpha=0.6, label='Train')
# plt.hist(test_error, bins=50, alpha=0.6, label='Test')
# plt.legend()
# plt.title("Reconstruction Error Distribution")
# plt.show()
#----------------실습2-----------------
# 시각화할 이미지 수
n=3

# 출력 그래프 크기 설정
plt.figure(figsize=(10,6))

# 노이즈 이미지 원본 복원 이미지 순서로 출력
for i in range(1,n+1):

  # 첫번째 행 노이즈 이미지 시각화
  ax=plt.subplot(3,n,i)
  plt.imshow(noisy_test[i].reshape(28,28), cmap='gray')

  # 두번째 행 원본 이미지 시각화
  ax=plt.subplot(3,n,i+n)
  plt.imshow(x_test[i].reshape(28,28), cmap='gray')

  # 세번째 행 복원된 이미지 시각화
  ax=plt.subplot(3,n,i+2*n)
  plt.imshow(decoded_img[i].reshape(28,28),cmap='gray')

# 그래프 표시
plt.show()
```
